```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("nyc_weather.csv")
df
```

```python
df.info()
```

```python
z_scores = np.abs(stats.zscore(df['Temperature']))
z_scores
```

```python
threshold = 1.25
```

```python
outlier_indices = np.where(z_scores > threshold)
```

```python
df_detected = df.drop(outlier_indices[0])
```

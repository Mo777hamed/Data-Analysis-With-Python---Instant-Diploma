```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colors = ["#89CFF0", "#FF69B4", "#FFD700", "#7B68EE", "#FF4500",
          "#9370DB", "#32CD32", "#8A2BE2", "#FF6347", "#20B2AA",
          "#FF69B4", "#00CED1", "#FF7F50", "#7FFF00", "#DA70D6"]
```

```python
df = pd.read_csv("shopping_trends_updated.csv")
df.sample(5)
```

```python
plt.figure(figsize=(20,6))

counts=df['Gender'].value_counts()
explode=(0,0.1)

counts.plot(kind='pie', colors=colors, explode=explode, autopct= lambda x: '{:.0f}'.format(x*counts.sum()/100))

plt.title('Males Vs Females')
plt.xlabel('Gender', weight='bold', color='#2F0F5D', fontsize=15, labelpad=20)
plt.ylabel('Count', weight='bold', color='#2F0F5D', fontsize=15, labelpad=20)
plt.legend(labels=counts.index, loc='best')
plt.show()
```
